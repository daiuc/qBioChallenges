```r
nobel = read_csv('nobel_nominations.csv', col_names = c('field', 'year', 'name'))
```

# analysis

## 1. for Chemistry, who received the most nominations? 

```r
filter(nobel, field == 'Che') %>% 
    group_by(name) %>% 
    tally %>% 
    arrange(-n) %>% 
    head
```

## 2. find all reserchers who received nom in > 1 field

```r
select(nobel, -year) %>% 
    distinct %>%
    group_by(name) %>%
    add_count() %>%
    filter(n > 1) %>%
    pull(name) %>%
    unique
```

```r
select(nobel, -year) %>% 
    distinct %>%
    group_by(name) %>%
    add_count() %>%
    filter(n > 1) %>%
    arrange(name)
```

## take phy, which year had the largest number of nominees?

```r
filter(nobel, field == 'Phy') %>%
    distinct %>%
    group_by(year) %>%
    tally %>%
    arrange(-n) %>% head
```

## for each field, average number of nom per year?

```r
nobel %>% 
    group_by(field, year) %>%
    tally %>%
    group_by(field) %>%
    summarise(meanNomPerYear = mean(n))
```
